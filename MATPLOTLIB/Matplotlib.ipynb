# Figures and subplots


```python
x = np.linspace(1,10,100)
y = np.log(x)

plt.figure(1)
plt.subplot(121)# equivalent to plt.subplot(1,2,1)
plt.title("y = log(x)")
plt.plot(x,y)

plt.subplot(222)
plt.title("y = log(x)**2")#equivalent to plt.subplot(1,2,2)
plt.plot(x,y**2)
plt.show()
```

```python
# Example: Create a fugure having 4 subplots
x = np.linspace(1, 10, 100)

plt.figure(1)

plt.subplot(2,2,1)
plt.title("Linear")
plt.plot(x,x)

plt.subplot(2,2,2)
plt.title("Cubic")
plt.plot(x,x**3)

plt.figure(2)
plt.subplot(2,2,1)
plt.title("Exponential")
plt.plot(x, x**2)

plt.subplot(2,2,2)
plt.title("Log")
plt.plot(x,np.log(x))
```

Types of commonly known plots

```python
import matplotlib.pyplot as plt
import numpy as np

# Create a 4x4 grid of subplots
fig, axs = plt.subplots(4, 4, figsize=(12, 12))  # 4 rows, 4 columns

# Create data for plotting
x = np.linspace(0, 10, 100)

# Loop through each subplot and plot data
for i in range(4):  # Loop over rows
    for j in range(4):  # Loop over columns
        ax = axs[i, j]  # Access each subplot
        y = np.sin(x + i + j)  # Example: sinusoidal wave with different phase shifts
        ax.plot(x, y)
        ax.set_title(f"Plot ({i+1}, {j+1})")  # Set a title for each plot

# Adjust layout to prevent overlap
plt.tight_layout()

# Show the plots
plt.show()
```

```python
sns.jointplot(data = df,x='Humidity (%)',y='Temperature (°C)')
plt.show()
```

```python
df2 = df [((df['Humidity (%)'] <100) & (df['Temperature (°C)']>20))]
sns.jointplot(data = df2,x='Humidity (%)',y='Temperature (°C)')
plt.show()
```

```python
df2 = df [((df['Humidity (%)'] <100) & (df['Temperature (°C)']>20))]
sns.jointplot(data = df2,x='Humidity (%)',y='Temperature (°C)',kind ="hex")
plt.show()
```
