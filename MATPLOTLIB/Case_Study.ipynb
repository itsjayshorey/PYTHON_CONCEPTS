```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('file.csv')
df.head()
```

```python
sns.boxplot(x = 'Condition' , y = 'Humidity (%)',data = df)
plt.xticks(rotation = 90)
# plt.yscale('log')
plt.show()
```

```python
sns.boxplot(x = 'Condition' , y = 'Temperature (°C)',data = df)
plt.xticks(rotation = 90)
# plt.yscale('log')
plt.show()
```

```python
sns.barplot(x = 'Condition' , y = 'Humidity (%)',data = df)
plt.xticks(rotation = 90)
# plt.yscale('log')
plt.show()
```

```python
grouped_data = df.groupby('Condition')
```

```python
print(grouped_data['Temperature (°C)'].mean())
```

```python
for name, group in grouped_data:
    print(f"Group: {name}")
    print(group)
    print("\n")
```
