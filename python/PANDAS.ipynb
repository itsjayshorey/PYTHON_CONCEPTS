### Merging 

```python
d1 = pd.DataFrame({"city" : ["Lucknow","Kanpur","Agra","Delhi"], "Temperature" : [32,45,30,40]})
```

```python
d1
```

```python
d2 = pd.DataFrame({"city" : ["Delhi","Lucknow","Kanpur"], "Humidiity" : [32,45,40]})
```

```python
d2
```

```python
pd.merge(d1,d2,on='city')
```

```python
pd.merge(d1,d2, on = ['city'], how = 'outer')
```

```python
pd.merge(d1,d2, on = ['city'], how = 'inner')
```

```python
pd.merge(d1,d2, on = ['city'], how = 'left')
```

```python
pd.merge(d1,d2, on = ['city'], how = 'right')
```
